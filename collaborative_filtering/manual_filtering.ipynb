```python
import pandas as pd

# Load the dataset
data = pd.read_csv('C:/capstone_project/collaborative_filtering/dahdigabungdahinggris.csv')

# Function to filter places based on user input for category and price range
def recommend_places(category=None, min_price=None, max_price=None):
    filtered_data = data
    
    if category:
        filtered_data = filtered_data[filtered_data['category'].str.lower() == category.lower()]
    
    if min_price is not None:
        filtered_data = filtered_data[filtered_data['price'] >= min_price]
    
    if max_price is not None:
        filtered_data = filtered_data[filtered_data['price'] <= max_price]
    
    # Selecting relevant columns
    result = filtered_data[['place', 'description', 'category', 'price', 'latitude', 'longitude']]
    
    return result

# Main function to get user input and display recommendations
def main():
    print("Welcome to J-GO!")
    
    # Get user input for category
    category = input("Please enter the category you are interested in (e.g., Taman Hiburan, Budaya): ")
    
    # Get user input for minimum price
    min_price_input = input("Please enter your minimum price (e.g., 0): ")
    min_price = int(min_price_input) if min_price_input.isdigit() else None
    
    # Get user input for maximum price
    max_price_input = input("Please enter your maximum price (e.g., 15000): ")
    max_price = int(max_price_input) if max_price_input.isdigit() else None
    
    # Get recommended places
    recommended_places = recommend_places(category=category, min_price=min_price, max_price=max_price)
    
    # Display results
    if not recommended_places.empty:
        print("\nRecommended Places:")
        print(recommended_places)
    else:
        print("No places found matching your criteria.")

# Run the main function
if __name__ == "__main__":
    main()
```
